```python
import matplotlib.pyplot as plt 

x = [1,7,3]
y = [9,4,5]

plt.plot(x,y)

plt.xlabel("Plot number")
plt.ylabel("Important var")

plt.title("Interesting table\nCheck it out")

plt.show()
```

```python
import matplotlib.pyplot as plt 

x = [1,2,3]
y = [4,58,6]

x2  = [66,22,39]
y2 = [98,29,43]

plt.plot(x,y, label = "First line")
plt.plot(x2,y2,label = "second line")


plt.xlabel("Plot number")
plt.ylabel("Important var")

plt.title("Interesting table\nCheck it out")
plt.legend()
plt.show()
```
