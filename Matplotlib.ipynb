# BAR CHARTS AND HISTOGRAMS

```python
import matplotlib.pyplot as plt

x = [2,4,6,8,10]
y = [6,7,8,2,4]

x2 = [3,8,24,18,10]
y2 = [6,7,8,2,4]

plt.bar(x,y, label = "bar1",color="r" )
plt.bar(x2,y2, label = "bar2",color="b" )

plt.xlabel("x")
plt.ylabel("y")

plt.title("Interesting Graph")
plt.legend()
plt.show()
```

# Stack Plots 

```python
import matplotlib.pyplot as plt 
days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing  = [8,5,7,8,13]

plt.stackplot(days, sleeping, eating, working, playing, colors=["m","c","r","k"])

plt.xlabel("x")
plt.ylabel("y")
plt.title("Interesting Graph")
plt.legend()
plt.show()
```
